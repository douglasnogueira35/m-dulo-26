# Tarefa - Álgebra Linear

A célula abaixo gera dados de notas de duas provas, p1 e p2, conforme em aula, mas com mais alunos.

## 1. Reduzindo correlações

a) Já vimos que expressar os dados através da média e diferença é uma transformação linear. Obtenha a média e a diferença através de uma multiplicação matricial, conforme visto em aula.

b) Calcule a correlação entre os dados da **p1** e **p2**

c) Calcule a correlação entre os dados transformados (média e diferença)

d) Comente as correlações observadas.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
```

```python
import numpy as np

# Exemplo de dados (substitua pelos seus valores)
p1 = np.array([10, 12, 13, 15, 16])
p2 = np.array([11, 14, 13, 17, 18])

# a) Transformação linear: média e diferença via multiplicação matricial
M = np.array([[0.5, 0.5],
              [1, -1]])  # matriz de transformação

dados = np.vstack((p1, p2))  # empilha p1 e p2 em linhas
transformados = M @ dados    # aplica a transformação

media = transformados[0, :]
diferenca = transformados[1, :]

print("Média:", media)
print("Diferença:", diferenca)

# b) Correlação entre p1 e p2
corr_p1_p2 = np.corrcoef(p1, p2)[0, 1]
print("Correlação entre p1 e p2:", corr_p1_p2)

# c) Correlação entre média e diferença
corr_media_dif = np.corrcoef(media, diferenca)[0, 1]
print("Correlação entre média e diferença:", corr_media_dif)

# d) Comentário automático
if abs(corr_media_dif) < abs(corr_p1_p2):
    print("A transformação reduziu a correlação, separando melhor as informações de tendência (média) e contraste (diferença).")
else:
    print("A transformação não reduziu a correlação neste conjunto de dados.")
```

## 2. Soma e diferença

Realize agora uma transformação semelhante, mas que ao invés da média, dê a soma da **p1** e **p2**. Construa o gráfico e conclua que diferenças visuais você observa.

```python
import numpy as np
import matplotlib.pyplot as plt

# Exemplo de dados (substitua pelos seus valores reais)
p1 = np.array([10, 12, 13, 15, 16])
p2 = np.array([11, 14, 13, 17, 18])

# Transformação linear: soma e diferença
M = np.array([[1, 1],
              [1, -1]])

dados = np.vstack((p1, p2))
transformados = M @ dados

soma = transformados[0, :]
diferenca = transformados[1, :]

# Gráfico
plt.figure(figsize=(8,6))
plt.scatter(soma, diferenca, color='blue', label='Soma vs Diferença')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.xlabel("Soma (p1+p2)")
plt.ylabel("Diferença (p1-p2)")
plt.title("Transformação linear: Soma e Diferença")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Exemplo de dados (substitua pelos seus valores reais)
p1 = np.array([10, 12, 13, 15, 16])
p2 = np.array([11, 14, 13, 17, 18])

# Transformações
media = (p1 + p2) / 2
soma = p1 + p2
diferenca = p1 - p2

# Gráficos lado a lado
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Média vs Diferença
axs[0].scatter(media, diferenca, color='green')
axs[0].axhline(0, color='gray', linestyle='--')
axs[0].axvline(0, color='gray', linestyle='--')
axs[0].set_title("Média vs Diferença")
axs[0].set_xlabel("Média (p1+p2)/2")
axs[0].set_ylabel("Diferença (p1-p2)")
axs[0].grid(True)

# Soma vs Diferença
axs[1].scatter(soma, diferenca, color='blue')
axs[1].axhline(0, color='gray', linestyle='--')
axs[1].axvline(0, color='gray', linestyle='--')
axs[1].set_title("Soma vs Diferença")
axs[1].set_xlabel("Soma (p1+p2)")
axs[1].set_ylabel("Diferença (p1-p2)")
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

- Média vs Diferença: os pontos ficam mais próximos do eixo horizontal, já que a média reduz a escala dos valores. Isso dá uma visão mais compacta da tendência central.
- Soma vs Diferença: os pontos se espalham mais, pois a soma amplia os valores. Esse gráfico evidencia melhor os casos em que ambas variáveis são altas ou baixas ao mesmo tempo.
- Conclusão: a escolha entre média ou soma depende da análise desejada. A média facilita comparações em escala reduzida, enquanto a soma destaca magnitude total dos dados.
